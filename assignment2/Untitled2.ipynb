```python
import random
DIRECTIONS = ['U', 'D', 'L', 'R']

# Function to create an individual (a random sequence of moves)
def create_individual(max_length):
    return ''.join(random.choice(DIRECTIONS) for _ in range(max_length))

# Initialize population
def initialize_population(population_size, max_path_length):
    return [create_individual(max_path_length) for _ in range(population_size)]

# Genetic algorithm to solve the maze
def genetic_algorithm(maze, start, end, population_size=100, generations=500, mutation_rate=0.01):
    max_path_length = len(maze) * len(maze[0])  # Maximum possible path length
    population = initialize_population(population_size, max_path_length)
    
    for generation in range(generations):
        # Evaluate fitness for each individual
        fitnesses = [fitness(ind, maze, start, end) for ind in population]
        
        # If a solution is found
        if 0 in fitnesses:
            return population[fitnesses.index(0)]
        
        # Selection
        selected = selection(population, fitnesses)
        
        # Crossover and mutation
        population = [mutate(crossover(selected[i], selected[i + 1]), mutation_rate) for i in range(0, len(selected) - 1, 2)]
    
    # Return the best solution found
    fitnesses = [fitness(ind, maze, start, end) for ind in population]
    return population[fitnesses.index(min(fitnesses))]

# Define start and end points
start = (0, 0)
end = (13, 24)

# Run the genetic algorithm
solution = genetic_algorithm(maze, start, end)
print("Solution found:", solution)
```

```python
# Fitness function to calculate how close a path gets to the end
def fitness(path, maze, start, end):
    x, y = start
    penalty = 0
    for move in path:
        if move == 'U' and x > 0:
            x -= 1
        elif move == 'D' and x < len(maze) - 1:
            x += 1
        elif move == 'L' and y > 0:
            y -= 1
        elif move == 'R' and y < len(maze[0]) - 1:
            y += 1
            # Check if the new position is a wall
        if maze[x][y] == 1000:
            penalty += 1000  # Penalize for hitting a wall
        elif (x, y) == end:
            return 0  # Path reaches the goal
 
    # Return the total weight, lower is better
    return penalty
```

```python
# Selection based on fitness (tournament selection)
def selection(population, fitnesses):
    selected = []
    for _ in range(len(population)):
        idx1, idx2 = random.sample(range(len(population)), 2)
        if fitnesses[idx1] < fitnesses[idx2]:
            selected.append(population[idx1])
        else:
            selected.append(population[idx2])
    return selected

# Crossover function
def crossover(parent1, parent2):
    point = random.randint(1, len(parent1) - 1)
    return parent1[:point] + parent2[point:]

# Mutation function to introduce randomness
def mutate(individual, mutation_rate=0.01):
    return ''.join(c if random.random() > mutation_rate else random.choice(DIRECTIONS) for c in individual)
```
