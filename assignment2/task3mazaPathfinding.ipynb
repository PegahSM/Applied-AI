```python
def fitness(path, maze, start, end):
    x, y = start
    weight = 0
    for move in path:
        # Check the current move and next position
        if move == 'U' and x > 1:
            if maze[x-1][y] == 1000:
                return float('inf')  # Infinite cost for invalid path
            weight += maze[x-1][y]
            x -= 2
        elif move == 'D' and x < len(maze)-2:
            if maze[x+1][y] == 1000:
                return float('inf')
            weight += maze[x+1][y]
            x += 2
        elif move == 'L' and y > 1:
            if maze[x][y-1] == 1000:
                return float('inf')
            weight += maze[x][y-1]
            y -= 2
        elif move == 'R' and y < len(maze[0])-2:
            if maze[x][y+1] == 1000:
                return float('inf')
            weight += maze[x][y+1]
            y += 2
    
    # Return the total weight, lower is better
    return weight
```

```python
def selection(population, fitnesses):
    tournament_size = 5
    selected = []
    for _ in range(len(population)):
        tournament = random.sample(range(len(population)), tournament_size)
        tournament_fitnesses = [fitnesses[i] for i in tournament]
        selected.append(population[tournament[tournament_fitnesses.index(max(tournament_fitnesses))]])
    return selected
```

```python
def crossover(parent1, parent2):
    point = random.randint(1, len(parent1)-1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2
```

```python
def mutate(individual, mutation_rate):
    return ''.join(c if random.random() > mutation_rate else random.choice(DIRECTIONS) for c in individual)
```

```python
def genetic_algorithm(maze, start, end, population_size=100, generations=100, mutation_rate=0.1):
    path_length = len(maze) * len(maze[0])  # Maximum possible path length
    
    # Initialize population
    population = [create_individual(path_length) for _ in range(population_size)]
    
    for generation in range(generations):
        # Evaluate fitness
        fitnesses = [fitness(ind, maze, start, end) for ind in population]
        
        # Check if solution found
        best_fitness = min(fitnesses)
        best_individual = population[fitnesses.index(best_fitness)]
        print(f"Generation {generation}: Best Fitness = {best_fitness}")
        
        if start == end or best_fitness < 1000:
            return best_individual
        
        # Selection
        selected = selection(population, fitnesses)
        
        # Crossover
        offspring = []
        for i in range(0, len(selected), 2):
            if i+1 < len(selected):
                child1, child2 = crossover(selected[i], selected[i+1])
                offspring.extend([child1, child2])
            else:
                offspring.append(selected[i])
        
        # Mutation
        population = [mutate(ind, mutation_rate) for ind in offspring]
    
    # Return best solution found
    #fitnesses = [fitness(ind, maze, start, end) for ind in population]
    #return population[fitnesses.index(max(fitnesses))]
    return best_individual

# Run the algorithm
start = (0, 0)
end = (12, 36)
solution = genetic_algorithm(maze, start, end)
print("Solution found:", solution)
```
